```python
import numpy as np
import scipy as sp
from astropy.coordinates import SkyCoord
import astropy as ap
import matplotlib.pyplot as plt
from astropy import units as u
import pyfits

%matplotlib inline
```

```python
f = pyfits.open('1523376166.71_6.45080082827_62.7257022297.fits')
```

```python
f[0].header
```

```python
freq = np.linspace(-12,12,8192)
```

```python
plt.figure(figsize = (10,6))
for i in np.arange(1,51):
    plt.plot(freq, f[i].data['auto0_real'])
    plt.title('first polarization', fontsize = 25)
    plt.xlabel('frequency [MHz]', fontsize = 22)
    plt.ylabel('Spectrum', fontsize = 22)
    plt.tick_params(labelsize = 14)
```

```python
plt.figure(figsize = (10,6))
for i in np.arange(1,51):
    plt.plot(freq, f[i].data['auto1_real'])
    plt.title('second polarization', fontsize = 25)
    plt.xlabel('frequency [MHz]', fontsize = 22)
    plt.ylabel('Spectrum', fontsize = 22)
    plt.tick_params(labelsize = 14)
```
