```python
def power(compData, div, N):
    vsamp = 62.5
    time = np.linspace(-N/2./vsamp, (N/2. - 1)/vsamp, N)
    freq = np.linspace(-vsamp/2, (vsamp/2)*(1-2./N), N)
    fft_ = np.fft.fft(compData[:,:N])
    power = abs(fft_)**2
    return freq, power
```
