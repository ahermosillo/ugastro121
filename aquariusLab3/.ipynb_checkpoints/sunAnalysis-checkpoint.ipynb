```python
import numpy as np
import astropy as ap
from astropy import units as u
import matplotlib.pyplot as plt
import ugradio
import time

%pylab inline
```

```python
vv = np.load('Sun_data_last.npz')['vv'][:-1000]
tt = np.load('Sun_data_last.npz')['ttt'][:-1000]

jdt = (ugradio.timing.julian_date(tt) - ugradio.timing.julian_date(tt)[0])*24
```

```python
plt.figure(figsize = (10,6))
plt.plot(jdt, vv, 'k')
plt.xlim(-1,11)
plt.grid()
plt.xlabel('Hours of Recording')
plt.ylabel('Volts')
```

```python
ff = np.fft.fft(vv)
power = np.fft.fftshift(abs(ff)**2)
freq = np.fft.fftshift(np.fft.fftfreq(len(jdt), d = 0.5))
plt.plot(freq,power)
plt.ylim(0,60)
plt.xlim(-0.02, 0.02)
```
