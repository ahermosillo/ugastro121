```python
for i in range(40):
    x = np.load('CasA_{}.npz'.format(i))['tt']
    y = np.load('CasA_{}.npz'.format(i))['v']
    plt.plot(x,y)
```

```python
VOLTZ = []
TIMEZ = []
for i in range(40):
    x = np.load('CasA_{}.npz'.format(i))['tt']
    y = np.load('CasA_{}.npz'.format(i))['v']
    
    VOLTZ = np.append(VOLTZ,y)
    TIMEZ = np.append(TIMEZ,x)

    

    

```

```python
plt.plot(TIMEZ,VOLTZ,'.',markersize=.2)
plt.show()
```

```python
plt.hist(VOLTZ,bins=30)
```

```python
PS = np.abs(np.fft.fft(VOLTZ))**2
freq = np.fft.fftfreq(len(VOLTZ),d=0.5)
```

```python
plt.semilogy(np.fft.fftshift(freq),np.fft.fftshift(PS))

plt.show()
```

```python
plt.plot(np.fft.fftshift(freq),np.fft.fftshift(PS))
plt.ylim(0,0.001)
plt.xlim(-0.1,0.1)
plt.show()
```
