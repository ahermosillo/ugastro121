```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 11,300)

xs = np.linspace(-1, 11,300) 

f = sinx(0.2*x)
g = sinx(xs)
h = sinx(0.9*xs)
plt.figure(figsize = (11,3))
plt.plot(x, f, 'b--', label = 'undersampled')
plt.plot(xs, g, 'k-')
plt.plot(xs, h, 'g--', label = 'bettersampled')

idx = np.argwhere(np.diff(np.sign(h - g)) != 0).reshape(-1) + 0
plt.plot(x[idx], h[idx], 'o')

idx2 = np.argwhere(np.diff(np.sign(f - g)) != 0).reshape(-1) + 0
plt.plot(xs[idx2], f[idx2], 'ko')
plt.tick_params(labelsize = 15)
plt.legend(bbox_to_anchor=(0.935, 0.4), )
plt.savefig('aliasing', format = 'pdf')
```
